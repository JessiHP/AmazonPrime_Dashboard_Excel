### Análisis bivariado de las variables numéricas

Tras realizar un análisis univariante sobre las columnas numéricas, profundizamos en la comparación de estas variables con la columna objetivo "Churn Label". De esta forma, buscamos extraer información valiosa sobre la relación entre las variables numéricas y la tasa de abandono (Churn Label). Para ello, creamos tablas y gráficos dinámicos.

``- Account length (months):``Observamos que cuanto más antigua es la cuenta del suscriptor, menor es la tasa de abandono (menor probabilidad de cancelación). Además, los suscriptores con una antigüedad de entre 1 y 5 meses tienen más probabilidades de cancelar la suscripción que de continuar en el servicio. 

>Propuesta: promover algún tipo de plan de beneficios para los clientes fieles y así fomentar su permanencia.



``- Customer Service Calls:`` A mayor número de llamadas al servicio de atención al cliente, mayor es la probabilidad de cancelar la suscripción. Es especialmente significativo el cambio de patrón cuando los clientes llaman más de tres veces: la probabilidad de cancelar la suscripción aumenta un 50% y mantener la suscripción se convierte en el escenario menos probable. Por lo tanto, una vez que el cliente llama al servicio, es fundamental que el problema se resuelva en un máximo de dos llamadas, momento en el que es muy probable que el cliente se moleste y tienda a cancelar la suscripción. ``Relevante para nuestro analisis``



``- Average Monthly GB Download:`` no existe una relación significativa entre los gigabytes descargados al mes y la probabilidad de cancelar la suscripción. En otras palabras, los clientes que no utilizan esta función tienen la misma probabilidad de cancelar la suscripción que los clientes que descargan grandes cantidades de gigabytes.




``- Extra Data Charges:`` la gran mayoría de los suscriptores (aprox. el 90%) no tienen cargos adicionales por datos. Por lo tanto, en este caso, el único análisis interesante sería centrarse en el 10% restante. Tras el filtrado, observamos que existe una relación significativa entre las variables; sin embargo, no hay suficientes datos por segmento para establecer conclusiones sólidas en este caso particular.



``- Age:`` las tasas de abandono se mantienen constantes (aprox. el 24%) hasta que los suscriptores alcanzan la tercera edad, momento en el que la probabilidad de cancelar la suscripción comienza a aumentar significativamente. Esto podría estar relacionado con dos problemas: la mayoría del contenido audiovisual ofrecido no es adecuado para personas mayores o la plataforma carece de accesibilidad para este grupo de edad. 

>Propuesta: Sería interesante investigar más a fondo el motivo por el que las personas mayores cancelan su suscripción y, quizás, lanzar una campaña de marketing específica para personas mayores aplicando estrategias para reducir sus tasas de abandono.



``- Number of customers in Group:`` casi el 80% de los suscriptores pagan un Plan Individual. Sin embargo, este es el plan con mayor tasa de abandono (32,76%). La probabilidad de cancelar la suscripción disminuye drásticamente (hasta aproximadamente un 6%) si el suscriptor paga un Plan Familiar, independientemente de cuántos clientes lo integren. La principal razón de estos patrones podría estar relacionada con la concesión que sienten los suscriptores al pagar un Plan Familiar, prefiriendo mantener la suscripción si alguien del grupo desea seguir disfrutando del servicio. ``Relevante para nuestro analisis``

>Propuesta: Fomentar los Planes Familiares en lugar de los planes individuales para reducir las tasas de abandono generales. 




``- Monthly Charges (USD):`` Cuanto mayores sean los cargos mensuales, mayor será la probabilidad de cancelación de la suscripción (esta regla no se aplica a los dos últimos segmentos; sin embargo, estos segmentos representan una pequeña proporción del total de suscriptores). Los clientes que pagan cantidades mayores pueden controlar mejor sus gastos y evaluar con mayor profundidad la relación calidad-precio, optando finalmente por alternativas. ``Relevante para nuestro analisis``

>Propuesta: Investigar si existen beneficios significativos para quienes pagan tales cantidades por el servicio y evaluar si realmente valen la pena.





``- Total charges (USD):`` Observamos que cuanto mayores sean los cargos totales (USD), menor será la probabilidad de cancelación de la suscripción. Sin embargo, esto probablemente esté relacionado con el tiempo que los clientes llevan pagando la suscripción: a mayor antigüedad de la cuenta, mayores serán los cargos acumulados y menores las tasas de abandono. 




``- Average Monthly Expenses (USD):`` Esta columna se comporta de forma muy similar a la variable Gastos mensuales y se aplican los mismos datos.


``- Number of Complaints or Support:`` Sorprendentemente, los clientes sin quejas ni solicitudes de soporte registradas muestran tasas de abandono similares a las de aquellos entre 1 y 10 solicitudes. Por el contrario, el número de llamadas de servicio al cliente se correlaciona positivamente con la tasa de abandono.

Esta discrepancia puede deberse a la naturaleza de los problemas. Es probable que los clientes que llaman al servicio de atención al cliente se enfrenten a problemas críticos que requieren una resolución inmediata o que hayan alcanzado un mayor nivel de insatisfacción.

Si estos problemas siguen sin resolverse después de varios intentos, es más probable que los clientes cancelen sus membresías.

Por otro lado, los clientes que presentan quejas o solicitudes de soporte pueden estar enfrentando problemas menos urgentes o aún no han alcanzado el umbral de frustración que les motiva a llamar. Además, la plataforma de quejas/soporte puede ofrecer un proceso de resolución más eficaz y satisfactorio que las llamadas de atención al cliente.

>Propuesta: Evaluar la eficacia del proceso de atención al cliente, especialmente en la resolución de problemas en las dos primeras llamadas, ya que los problemas sin resolver son un factor clave de la pérdida de clientes. También puede ser beneficioso analizar las fortalezas de la plataforma de quejas/soporte para identificar prácticas que podrían aplicarse a las llamadas de atención al cliente.
